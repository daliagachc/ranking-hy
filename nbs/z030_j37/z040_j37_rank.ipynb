```python

ser = (
    xr.merge([
        d37['j37']
        .rolling({'hour':5},min_periods=1).mean()
        .max('hour'),
        d5['d_lN']
    ])
    .squeeze()
    # .assign({'dN':lambda d:10**d['d_lN']})
    # .assign({'Q':lambda d:d['q']**3})
    .assign({'Q':lambda d:d['q']})
    .assign({'Qr':lambda d:(d['q']/5).round()*5})
    .assign({'lj37':lambda d:np.log10(d['j37'])})
    .to_dataframe()
    .where(lambda d:d['Q']>=0)
    .where(lambda d:d['j37']>0)
)
```
